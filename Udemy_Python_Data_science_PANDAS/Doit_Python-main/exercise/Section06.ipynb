## Question1 : 자동차 배기량에 따라 고속도로 연비가 다른지 알아보려고 함. displ(배기량)이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 hwy(고속도로 연비)평균이 더 높은지 알아보자.

### 방법 1 : 교재에 나온 방식, query 구문 이용

```python
# condition 1 : displ =< 4
condition1 = mpg.query('displ <= 4')
```

```python
# condition 2 : displ >= 5
condition2 = mpg.query('displ >= 5')
```

```python
condition1['hwy'].mean()
```

```python
condition2['hwy'].mean()
```

### 방법 2 : loc 메서드 이용, if문으로 결과 찾기

```python
condition1_mean = mpg.loc[mpg['displ'] <=4, 'hwy'].mean()
condition2_mean = mpg.loc[mpg['displ'] >= 5, 'hwy'].mean()
```

```python
# 평균값 비교
if condition1_mean > condition2_mean:
    result = "배기량이 4 이하인 자동차의 고속도로 연비 평균이 더 높습니다."
elif condition1_mean < condition2_mean:
    result = "배기량이 5 이상인 자동차의 고속도로 연비 평균이 더 높습니다."
else:
    result = "두 그룹의 고속도로 연비 평균이 같습니다."
```

```python
result
```

```python
if condition1_mean > condition2_mean:
    larger_value = condition1_mean
    result_text = f"배기량 4 이하의 고속도로 연비 평균이 더 높습니다: {round(condition1_mean)}"
elif condition1_mean < condition2_mean:
    larger_value = condition2_mean
    result_text = f"배기량 5 이상의 고속도로 연비 평균이 더 높습니다: {round(condition2_mean)}"
else:
    larger_value = condition1_mean  # 두 값이 같은 경우
    result_text = f"두 그룹의 고속도로 연비 평균이 동일합니다: {round(condition1_mean)}"
```

```python
result_text
```

## Question 2 : 자동차 제조 회사에 따라 도시 연비가 어떻게 다른지 알아보려고 함. 'audi'와 'toyota'중 어느 manufacturer의 cty 평균이 더 높은지 알아보자.

```python
mpg
```

```python
audi = mpg.loc[mpg['manufacturer'] == 'audi']
audi_cty_mean = audi['cty'].mean()

toyota = mpg.loc[mpg['manufacturer'] == 'toyota']
toyota_cty_mean = toyota['cty'].mean()
```

```python
if audi_cty_mean >= toyota_cty_mean:
    audi_win = f"아우디의 연비가 더 높습니다. 아우디의 연비 : {round(audi_cty_mean)}"
    print(audi_win)
else:
    toyota_win = f"토요타의 연비가 더 높습니다. 토요타의 연비 : {round(toyota_cty_mean)}"
    print(toyota_win)
```
