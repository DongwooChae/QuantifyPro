```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots=True,figsize=(15, 12), sharex=True

)
plt.show()
```

```python
titanic.plot(subplots=True,figsize=(15, 12), sharex=False, sharey=False

)
plt.show()
# sharey = true로 설정하면, 모든 서브플롯의 y축이 동일한 눈금으로 바뀌게 됨!
# fare열의 눈금이 가장 크므로, sharey는 이 fare열의 단위를 사용하게 됨
# 만약 age 처럼 특정한 열만 보고 싶다면?!
```

```python
titanic.age.plot(figsize=(12,8))
plt.show()
```

# customization

```python
titanic.age.plot(figsize=(12,8), fontsize= 13, c= "r", linestyle = "-", xlim=(0,900), ylim=(0,80),
                 xticks=xticks, yticks=yticks, rot=45)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend()
plt.xlabel("Passenger No", fontsize = 11)
plt.ylabel("Age", fontsize = 11)
plt.grid()
plt.show()


xticks = [x for x in range(0,901,50)]
yticks = [y for y in range(0,81,5)]
```
