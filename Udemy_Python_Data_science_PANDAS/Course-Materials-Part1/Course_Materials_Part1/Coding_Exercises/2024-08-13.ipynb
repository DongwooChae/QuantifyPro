# The agg() method
- 코드 한 줄로 원하는대로 요약 통계를 만드는 법

```python
titanic.age.mean()
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
# 예전에는 .mean, .agg 메소드를 그대로 이용할 수 있었으나 이제는 문자열이 있는 열의 값을 무시하고 계산이 불가능함. 따라서 select_dtypes() 메소드를 이용해서 수치형 데이터만 선택하고 .mean을 호출해야 함
titanic.age.mean()
titanic.age.agg("mean")
```

```python
titanic.age.agg(["mean", "std"])
```

```python
titanic.agg({"survived":"mean", "age":["min","max","mean"]})
```
