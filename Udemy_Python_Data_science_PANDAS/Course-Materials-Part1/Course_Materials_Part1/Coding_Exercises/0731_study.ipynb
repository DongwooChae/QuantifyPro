# 하나의 원소가 아니라, 행이나 열 안에서 여러 개의 원소를 한번에 바꾸는 법

## loc를 사용하여 single value 교체

```python
titanic.loc[1, "age"] = 40
# loc에서 인덱스, 열 지정
# 인덱스는 순서로, 열은 직접 칼럼명을 입력하였음
```

```python
titanic.head()
```

## iloc을 이용하여 데이터 입력하기
- iloc은 인덱스, 열 모두 숫자 위치값으로 지정하도록 함
- 따라서 loc에서 [1, 3]이라고 입력하면 1번째 인덱스의 값, '3'이라는 열을 생성하여 데이터를 입력하게 됨
- iloc을 이용해야 [1, 3]으로 위치 지정이 가능

```python
titanic.iloc[1, 3] = 41
# iloc에서 인덱스, 열 모두 순서로 지정하였음
# 만약 41이라는 숫자가 아니라 변수라면?
# 평가의 결과값을 저장해두는 데이터프레임을 만든 다음, 그 열에 입력할 값을 변수로 지정해놓고 투입시킨다면?
```

```python
titanic.head()
```

## 특정 열에 대해 여러 인덱스의 값을 고치고 싶다면?

```python
# loc 사용
titanic.loc[1:3, "age"] = 42
```

```python
titanic.head()
```

```python
# iloc 사용

# titanic.iloc[1:3, 3] = [43, 44, 45]
titanic.iloc[1:4, 3] = [43,44,45]

# iloc에서는 1:4로 입력할 경우 1~3 인덱스만 할당(4는 제외)되므로 3개의 인덱스에 대응하는 3개의 값이 리스트에 포함되어야 함
```

# Boolean indexing

```python
index_babies = titanic.loc[titanic.age < 1, "age"].index
```

```python
index_babies
```

```python
titanic.loc[index_babies]
```

```python
index_male = titanic.loc[titanic.sex == "male", "sex"].index
```

```python
index_male_age = titanic.loc[titanic.age.between(20,30), "age"].index
```

```python
titanic.loc[index_male]
# 여기서는 sex 열의 값이 male에 해당하는 것만 불리언 인덱싱으로 포함하였는데, 조건 여러 개를 같은 방식으로 걸고 싶다면?
```

```python
titanic_new_condition = titanic.loc[(index_male) & (index_male_age)]
```

```python
# 인덱스 객체가 아닌 불리언 인덱스로 변경하기
boolean_index_male = titanic.sex == "male"
boolean_index_age = titanic.age.between(20,30)
```

```python
test = titanic.loc[boolean_index_age & boolean_index_male]
```
