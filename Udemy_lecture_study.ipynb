# 멀티 인덱스(다중 인덱스 or 계층적 인덱스)
- 인덱스는 하나 이상의 열을 설정할 수 있음

## 다중 인덱스의 장점 : 데이터를 고차원 구조로 정리할 수 있음
- ex) groupby 메소드와 함께 데이터를 그룹지어 연산할 때 유용함

```python
titanic_multiindex = titanic.set_index(["pclass", "sex"])
# 별도의 변수를 설정하지 않고, 기존 변수의 데이터를 덮어쓰기 하고 싶다면?
# titanic.set_index(["pclass", "sex"], inplace = True) 즉, inplace = True를 하면 됨
# set_index() 메소드의 keys 파라미터에 여러개의 열을 넣으면 됨
```

```python
titanic_multiindex
# 현재 새 데이터프레임에서 다중 인덱스가 정렬되지 않았음. sort_index 메소드를 이용해 정렬했을 때 더 유용함
```

```python
# titanic_multiindex.sort_index()
# ascending 파라미터를 각각 다르게 설정하고 싶다면? >> 리스트를 만들어서 순서를 부여하면 됨
# pclass는 내림차순, sex는 오름차순으로

titanic_multiindex.sort_index(ascending= [False, True], inplace = True)
# 원하는대로 정렬이 됐다면, inplace = Ture로 덮어씌우면 됨
```

```python
titanic_multiindex
```

## swaplevel() 메소드
- 앞에서는 pclass , sex 순서로 인덱스를 정렬했는데, 반대로 성별 > 클래스 순으로 나타내고 싶다면?

# 다중 인덱스 2
- 다중 인덱싱을 이용한 인덱싱과 슬라이싱

```python
titanic = titanic.iloc[:50, ]
titanic
```

```python
titanic = titanic.set_index(["pclass","sex"]).sort_index(ascending=True)
# 여러 개의 메소드를 쭉 이어서 쓸 수도 있음. >> .set_index().sort_index()
```

```python
# 여기서 라벨 기반으로 데이터프레임을 인덱싱, 슬라이싱해보자.
# loc 연산자를 이용해 라벨 기반 인덱싱
# pclass 인덱스의 값이 1인 승객을 모두 선택하기 위해서는, 라벨인 1을 대괄호에 입력한 다음 선택할 수 있음
titanic.loc[1]
```

```python
# pclass의 값이 1 또는 2인 값을 모두 인덱싱하려면?
titanic.loc[[1,2]]
```

```python
# 1차, 2차 인덱스를 동시에 이용해서 데이터프레임을 슬라이스하기
```

```python
titanic.loc[([1,2], "female"), ["age", "fare"]]
```

```python
titanic.loc[([1,2], "female"), : ]
```

```python
# titanic.loc[([1,2,3] ,"female"), :]
# 2차 인덱스를 이용해서 데이터프레임을 슬라이스하기
# 2차 인덱스가 female인 모든 행을 선택하자
# 1차 인덱스는 1, 2, 3 모든 값을 가지고 있어야 한다.
# 직관적으로 생각해보면 loc 연산자를 사용해서 titanic.loc[(:,"female"), : ]과 같이 pclass의 모든 값과 sex의 female값, 모든 열을 선택하면 될 것으로 보이지만, 실행이 되지 않는다.
# 이때 행 라벨을 단순히 (:, "female")로 인덱싱하면 안되고,
# 명시적으로 1차 인덱스를 슬라이스하지 않을 것임을 표시해주기 위해 slice 키워드를 이용해야 한다.

# titanic.loc[(slice(None), slice("female")), :]과 같이 1차인덱스는 슬라이스 하지 않고, 2차 인덱스는 "female"행을 슬라이스 한 뒤 모든 열을 선택하고 실행하자.
titanic.loc[(slice(None), slice("female")), :]
```
