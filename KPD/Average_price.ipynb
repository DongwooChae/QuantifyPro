```python
# Raw_data 추출

industry_url = 'https://github.com/DongwooChae/QuantifyPro/blob/master/KPD/00.%ED%91%9C%EC%A4%80%EC%82%B0%EC%97%85%EB%B6%84%EB%A5%98.xlsx?raw=true'
price_url = 'https://github.com/DongwooChae/QuantifyPro/blob/master/KPD/01_%EC%A3%BC%EA%B0%80.xlsx?raw=true'
value_per_stock_url = 'https://github.com/DongwooChae/QuantifyPro/blob/master/KPD/02_%EC%A3%BC%EB%8B%B9%EA%B0%80%EC%B9%98.xlsx?raw=true'
fsdata_url = 'https://github.com/DongwooChae/QuantifyPro/blob/master/KPD/03_%EC%9E%AC%EB%AC%B4%EB%8D%B0%EC%9D%B4%ED%84%B0.xlsx?raw=true'

industry = pd.read_excel(industry_url, index_col = '종목명')
price = pd.read_excel(price_url, parse_dates=['Date'], index_col='Date')
value_per_stock = pd.read_excel(value_per_stock_url)
fsdata = pd.read_excel(fsdata_url)
```

# 0. 표준산업분류 업종으로 검색, 해당 기업 저장하기

```python
industry_keyword_input = input("검색할 표준산업분류 업종 키워드를 입력하세요 (예: 게임, 소프트웨어, 기타 서비스): ")
industry_keyword = [k.strip() for k in industry_keyword_input.split(",")]
# 키워드 리스트를 OR 연산자로 연결하여 정규표현식 패턴을 만듭니다.
pattern = '|'.join(industry_keyword)


# '10차표준(세분류)' 열에서 해당 패턴을 대소문자 구분 없이 검색하여, 
# 조건에 맞는 행의 인덱스를 추출합니다.
matching_indices = industry[industry['10차표준(세분류)'].astype(str)
                              .str.contains(pattern, case=False, na=False)].index

# industry.head(5)
```

```python
price_sample = price.loc[:,price.columns.isin(matching_indices)]
```

# 1. 시작 날짜 및 종료 일자 입력

## 1.1 평균값을 계산할 범위 입력(ex: 2024-06-30 ~ 2024-10-31)

```python
start_date_str = input("시작 날짜를 입력하세요. (YYYY-MM-DD): ")
start_date = pd.to_datetime(start_date_str)

# 시작 날짜로부터 6개월 후 계산(pd.DateOffset 사용)
end_date_str = input("종료 날짜를 입력하세요. (YYYY-MM-DD): ")
end_date = pd.to_datetime(end_date_str)

# pd.date_rage로 DatetimeIndex 생성(일 단위로 생성, freq='D')
date_index = pd.date_range(start=start_date, end=end_date, freq='D')
```

## 1.2 해당 날짜 범위 필터링하여 price 데이터프레임 교집합 산출

```python
# filtered_price_sample = price_sample.loc[price_sample.index.intersection(date_index)]
filtered_price = price_sample.loc[price_sample.index.intersection(date_index)]
```
