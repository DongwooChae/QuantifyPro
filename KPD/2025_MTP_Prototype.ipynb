```python
# 정렬된 값을 사용하여 체크박스 생성
checkboxes = [widgets.Checkbox(value=False, description=val) for val in sorted_values]

# 1행당 5개씩 배치하기 위해 리스트를 분할하여 HBox에 담기
rows = []
row_length = 5
for i in range(0, len(checkboxes), row_length):
    row = widgets.HBox(checkboxes[i:i+row_length])
    rows.append(row)
checkbox_container = widgets.VBox(rows)
```

```python
# 결과 출력용 위젯 생성
output = widgets.Output()
```

```python
# 확인 버튼과 리셋 버튼 생성
confirm_button = widgets.Button(description='확인')
reset_button = widgets.Button(description='리셋')
```

```python
def on_confirm_button_clicked(b):
    output.clear_output()  # 이전 출력 지우기
    # 체크된 항목의 설명(카테고리)을 가져옴
    selected_categories = [cb.description for cb in checkboxes if cb.value]
    
    with output:
        if not selected_categories:
            print("선택된 항목이 없습니다.")
        else:
            # 선택된 카테고리에 해당하는 종목명(인덱스) 수집
            selected_indices = []
            for cat in selected_categories:
                indices = industry[industry['10차표준(세분류)'].astype(str) == cat].index.tolist()
                selected_indices.extend(indices)
            selected_indices = list(set(selected_indices))
            
            # value_per_stock에서 해당 종목명의 EPS, SPS, BPS 열 가져오기
            try:
                selected_data = value_per_stock.loc[selected_indices, ['EPS', 'SPS', 'BPS']]
                display(selected_data)
            except Exception as e:
                print("데이터를 가져오는 도중 에러 발생:", e)

def on_reset_button_clicked(b):
    # 모든 체크박스 선택 해제
    for cb in checkboxes:
        cb.value = False
    output.clear_output()  # 출력창 초기화
    with output:
        print("체크박스가 초기화되었습니다. 선택 후 확인 버튼을 눌러주세요.")



# 중복 등록 방지: 기존 이벤트 핸들러 제거 후 등록
confirm_button.unobserve_all()
confirm_button.on_click(on_confirm_button_clicked)

reset_button.unobserve_all()
reset_button.on_click(on_reset_button_clicked)
```

```python
# 버튼들을 포함하여 화면에 표시 (원하는 순서대로 배치)
display(checkbox_container, widgets.HBox([confirm_button, reset_button]), output)
```
