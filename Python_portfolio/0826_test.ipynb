```python
meta = yf.download('META', start="2024-01-01", end="2024-07-31")
meta.head()
```

```python
# 'Date' 열이 제대로 존재하는지 확인
print(meta.columns)

# 인덱스가 'Date'인 경우에는 reset_index 후 이름이 변환된 열을 확인
meta = meta.reset_index()

# 열 이름이 Date인지 다시 확인
if 'Date' not in meta.columns:
    print(meta.columns)  # 열 이름 출력해서 확인

# 'Date'가 있는 경우 필터링 수행
if 'Date' in meta.columns:
    meta = meta[meta['Date'] >= '2024-01-01']

fig = go.Figure(data=[go.Candlestick(x=meta['Date'],
                open=meta['Open'],
                high=meta['High'],
                low=meta['Low'],
                close=meta['Close'])])
fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

```python
stock = yf.download(tickers['Symbol'].tolist(), start="2023-01-01-", end="2024-07-31")
stock
```

```python
stock = stock.reset_index()
starts = stock[stock['Date'] == '2023-01-03']['Open'].transpose()
starts = starts.reset_index()
ends = stock[stock['Date']=='2024-07-31']['Close'].transpose()
ends = ends.reset_index()

starts.columns = ['Ticker', 'START']
ends.columns = ['Ticker', 'END']

stocks_ = pd.merge(starts, ends, how='left', left_on='Ticker', right_on='Ticker')
stocks_['rate'] = (stocks_['END']- stocks_['START']) / stocks_['START']
stocks_ = stocks_.sort_values('rate', ascending=False)
stocks_
```

```python
# starts와 ends의 구조를 확인
print(starts.head())
print(ends.head())

# 만약 'Date'를 인덱스로 설정했다면 다시 설정 해제 후 적절한 열 이름을 설정해야 합니다.
# 예시: starts와 ends에 열이 하나밖에 없을 경우
starts = stock[stock['Date'] == '2023-01-03'][['Ticker', 'Open']].transpose().reset_index()
ends = stock[stock['Date'] == '2024-07-31'][['Ticker', 'Close']].transpose().reset_index()

# 열 이름을 적절히 설정
starts.columns = ['Ticker', 'START']
ends.columns = ['Ticker', 'END']

# 두 데이터프레임을 병합
stocks_ = pd.merge(starts, ends, how='left', left_on='Ticker', right_on='Ticker')

# 수익률 계산
stocks_['rate'] = (stocks_['END'] - stocks_['START']) / stocks_['START']

# 수익률 기준으로 정렬
stocks_ = stocks_.sort_values('rate', ascending=False)

# 결과 출력
print(stocks_)
```
