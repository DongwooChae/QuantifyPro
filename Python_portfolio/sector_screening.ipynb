# 주어진 티커에 대한 주주자본 가져오기

```python
# 빈 리스트로 데이터 수집
equity_data = []
```

```python
apple = yf.Ticker("AAPL")
apple_info = apple.info # 딕셔너리 형식

# dictionary to dataframe
apple_df = pd.DataFrame(list(apple_info.items()), columns=['Metric', 'Value'])

```

```python
sheet = wb.sheets['Sheet3']
# 컬럼 헤더 및 데이터프레임 데이터를 리스트로 변환하여 Excel에 출력

# NaN 값을 빈 문자열로 대체
apple_df.fillna('', inplace=True)
# 컬럼 헤더 및 데이터프레임 데이터를 리스트로 변환하여 Excel에 출력
sheet.range('A1').value = apple_df.columns.tolist()  # A1에 컬럼 헤더 출력
sheet.range('A2').value = apple_df.values.tolist()  # A2에 데이터 출력
```

```python
# 각 티커에 대해 주주자본 값을 가져오는 반복문
for ticker in tickers:
    stock = yf.Ticker(ticker)
    
    # 재무제표 데이터 가져오기
    balance_sheet = stock.balance_sheet
    
    # 최근 일자의 주주자본 (총 자본) 가져오기
    try:
        # "Total Stockholder Equity" 항목을 탐색하고 위치 기반으로 값을 가져옴
        equity = balance_sheet.loc['Stockholders Equity'].iloc[0]  # 최근 일자의 값을 가져옴
    except KeyError:
        # 값이 없을 경우 None으로 처리
        equity = None
    
    # 리스트에 딕셔너리 형태로 추가
    equity_data.append({'Ticker': ticker, 'Stockholders Equity': equity})

# 리스트를 데이터프레임으로 변환
equity_df = pd.DataFrame(equity_data)
```

```python
equity_df.to_csv('equity_data.csv', index=False)
```
